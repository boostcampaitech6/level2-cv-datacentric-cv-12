```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def create_spray_effect(image_size, num_dots, max_dot_radius, gap_probability):
    img = np.ones((image_size, image_size, 3), dtype=np.uint8) * 255

    #   
    center_x, center_y = image_size // 2, image_size // 2

    #    
    for _ in range(num_dots):
        x = np.random.randint(0, image_size)
        y = np.random.randint(0, image_size)
        #   
        distance_to_center = np.sqrt((x - center_x)**2 + (y - center_y)**2)
        while not distance_to_center < image_size//2:
            x = np.random.randint(0, image_size)
            y = np.random.randint(0, image_size)
            distance_to_center = np.sqrt((x - center_x)**2 + (y - center_y)**2)

        #    (  ,   )
        dot_radius = np.random.randint(1, max_dot_radius + 1)
        if distance_to_center > image_size / 3:
            #      
            dot_radius = int(dot_radius * (distance_to_center / (image_size / 2)))

        #      ( )
        if np.random.rand() > gap_probability:
            color = (0, 0, 0)  
            cv2.circle(img, (x, y), dot_radius, color, -1)

    return img

#  ,   ,   ,    
image_size = 50
num_dots = 500
max_dot_radius = 1  #   
gap_probability = 0  # 0.0 1.0  ,    

#    
spray_effect_img = create_spray_effect(image_size, num_dots, max_dot_radius, gap_probability)

#    
# plt.imshow(cv2.cvtColor(spray_effect_img, cv2.COLOR_BGR2RGB))
# plt.figure(figsize=(1, 1))
plt.imshow(cv2.cvtColor(spray_effect_img, cv2.COLOR_BGR2RGB), extent=[0, image_size/2, 0, image_size/2])
plt.axis('off')
plt.show()
```
